```python
import pandas as pd

df = pd.read_csv(
    "explainable-fake-news-detector/data/fake_news.csv",
    engine="python",
    on_bad_lines="skip"
)

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df['label'] = pd.to_numeric(df['label'], errors='coerce')

```

```python
df = df[df['label'].isin([0, 1])]
```

```python
df['label'] = df['label'].astype(int)
```

```python
df['label'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['text_length'] = df['text'].astype(str).apply(len)
df['text_length'].describe()
```

```python
df.groupby('label')['text_length'].mean()
```

```python
df[df['label'] == 1]['text'].sample(3, random_state=42)
```

```python
df[df['label'] == 0]['text'].sample(3, random_state=42)
```

```python
df = df.drop(columns=['text_length'])
```

```python
df = df.dropna(subset=['text'])
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv(
    "explainable-fake-news-detector/data/fake_news_cleaned.csv",
    index=False
)
```
